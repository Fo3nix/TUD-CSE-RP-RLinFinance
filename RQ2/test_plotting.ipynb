```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
filename = '.\logs\\20250522_144538_DQN_model.csv'
df = pd.read_csv(filename)
```

```python
df
```

```python
# only keep [timestep and info] 
df = df[['timestep', 'info']]
df
```

```python
from common.envs.forex_env import Snapshot, SimpleOHLC

# find 'close=' in info
df['equity'] = df['info'].apply(lambda x: x.split('close=')[1].split(' ')[0])
df['equity'] = df['equity'].apply(lambda x: x.split('))')[0])

#set snapshot as a floatS
df['equity'] = df['equity'].astype(float)
df
```

```python
# plot close per timestep
plt.plot(df['timestep'], df['equity'])
plt.xlabel('Timestep')
plt.ylabel('Equity')
plt.title('Equity over time')
plt.show()
```

```python
# show timestep 75000 to 100000
slice_df = df[(df['timestep'] >= 7500) & (df['timestep'] <= 10000)]
plt.plot(slice_df['timestep'], slice_df['equity'])
plt.xlabel('Timestep')
plt.ylabel('Equity')
plt.title('Equity over time')
plt.show()
```
